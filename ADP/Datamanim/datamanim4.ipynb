```python
import pandas as pd
```

```python
DataUrl='https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv'
df=pd.read_csv(DataUrl)
type(df)
```

```python
df.info()
```

```python
df.Yr_Mo_Dy=pd.to_datetime(df.Yr_Mo_Dy)
df.Yr_Mo_Dy
```

```python
df.Yr_Mo_Dy.dt.year.unique()
```

```python
def fixdate(x):
    import datetime

    year=x.year-100 if x.year>=2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df.Yr_Mo_Dy=df.Yr_Mo_Dy.apply(fixdate)
df.head()
```

```python
df.groupby(df.Yr_Mo_Dy.dt.year).mean().head()
```

```python
df['weeday']=df.Yr_Mo_Dy.dt.weekday
df.head()
```

```python
df.weeday.to_frame()
```

```python
def weekend (x):
    if x>=5:
        return 1
    else:
        return 0

df['WeekCheck']=df.weeday.apply(weekend)
df.WeekCheck.to_frame()
```

```python
df['WeekCheck']=df.weeday.map(lambda x:1 if x>=5 else 0)
df.WeekCheck.to_frame()
```

```python
df['WeekCheck']=df.weeday.map(lambda x:1 if x in [5,6] else 0)
df.WeekCheck.to_frame()
```
