```python
import pandas as pd
```

```python
Url='https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df=pd.read_csv(Url)
type(df)
```

```python
def fix_date(x):
    import datetime

    date=x.split(':')[0]
    hour=x.split(':')[1]

    if hour=='24':
        hour='00:00:00'
        final_date=pd.to_datetime(date+" "+hour) +datetime.timedelta(days=1)
    else:
        hour=hour+':00:00'
        final_date=pd.to_datetime(date+" "+hour)

    return final_date

df['(년-월-일:시)']=df['(년-월-일:시)'].apply(fix_date)
```

```python
def fixtime(x):
    import datetime

    date=x.split(':')[0]
    hour=x.split(':')[1]

    if hour=='24':
        hour='00:00:00'
        finaltime=pd.to_datetime(date+" "+hour)+datetime.timedelta(days=1)
    else:
        hour=hour+':00:00'
        finaltime=pd.to_datetime(date+" "+hour)
    return finaltime

df['(년-월-일:시)']=df['(년-월-일:시)'].apply(fixtime)
```
