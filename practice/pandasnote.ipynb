```python
df=sns.load_dataset('titanic')
```

```python
df.tail()
```

```python
df1=df.copy()
```

```python
df1['age'].mean()
```

```python
df1.loc[(df1.sex=='female')&(df1.age.isnull()), 'age']=df1.loc[df1.sex=='female', 'age'].mean()
df1.loc[(df1.sex=='male')&(df1.age.isnull()), 'age']=df1.loc[df1.sex=='male', 'age'].mean()
```

```python
df1['age'].mean()
```

```python
assert (df1['age'].isnull().sum() == 0)
assert df1['age'].mean().round(5) == 29.73603
```

```python
df1.tail()
```

```python
df1=df.copy()
```

```python
df1.loc[df1.sex=='female', 'age']=df1.loc[df1.sex=='female', 'age'].fillna(df1.loc[df1.sex=='female', 'age'].mean())
df1.loc[df1.sex=='male', 'age']=df1.loc[df1.sex=='male', 'age'].fillna(df1.loc[df1.sex=='male', 'age'].mean())
```

```python
df1.tail()
```

```python
assert (df1['age'].isnull().sum() == 0)
assert df1['age'].mean().round(5) == 29.73603
```

```python
df1=df.copy()
```

```python
df1.loc[df1.sex=='male', 'age']=df1.loc[df1.sex=='male', 'age'].fillna(df1.loc[df1.sex=='male', 'age'].mean())
df1.loc[df1.sex=='female', 'age']=df1.loc[df1.sex=='female', 'age'].fillna(df1.loc[df1.sex=='female', 'age'].mean())
```

```python
df1.tail()
```

```python
df1=df.copy()
```

```python
df1.loc[(df1.sex=='male')&(df1.age.isnull()), 'age']=df1.loc[df1.sex=='male', 'age'].mean()
df1.loc[(df1.sex=='female')&(df1.age.isnull()), 'age']=df1.loc[df1.sex=='female', 'age'].mean()
```
