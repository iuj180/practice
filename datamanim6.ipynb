```python
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv'
df=pd.read_csv(url)
type(df)
```

```python
df.head()
```

```python
df.drop('Indicator', axis=1, inplace=True)
df['First Tooltip']=df['First Tooltip'].map(lambda x: float(x.split("[")[0]))
df.head()
```

```python
df.loc[(df.Period>=2015)&(df.Dim1=='Both sexes')]
```

```python
df.loc[(df.Period>=2015)&(df.Dim1=='Both sexes')].pivot(index='Location', columns='Period', values='First Tooltip')
```

```python
df.pivot_table(index='Dim1', columns='Period', values='First Tooltip', aggfunc='mean').iloc[:,:4]
```
