```python
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
url1='https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv'
url2='https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv'
channel=pd.read_csv(url1)
video=pd.read_csv(url2)
type(channel)
type(video)
```

```python
display(channel)
display(video)
```

```python
channel.ct=pd.to_datetime(channel.ct)
video.ct=pd.to_datetime(video.ct)
```

```python
video.videoname.value_counts()
```

```python
video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep='last')[['viewcnt','videoname','ct']].reset_index(drop=True)
```

```python
channel.sort_values(['ct','channelname']).loc[channel.ct>=pd.to_datetime('2021-10-03')][['channelname','subcnt']].drop_duplicates('channelname').reset_index(drop=True)
```

```python
start=channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H')=='2021-10-03 03'][['channelname','subcnt']].reset_index(drop=True)
end=channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H')=='2021-11-01 15'][['channelname','subcnt']].reset_index(drop=True)
start.columns=['channelname','start_sub']
end.columns=['channelname','end_sub']
total=pd.merge(start,end)
total['delta']=total['end_sub']-total['start_sub']
total[['channelname','delta']]
```

```python
import datetime
EP_1=video.sort_values('ct').loc[video.videoname.str.contains('EP1')].reset_index(drop=True)
EP_1.loc[(EP_1.ct.diff(1)>=datetime.timedelta(minutes=20)) | (EP_1.ct.diff(1)<=datetime.timedelta(minutes=5))]
EP_1.loc[EP_1.index.isin([720,721,722,723,1635,1636,1637])]
```

```python
video['date']=video.ct.dt.strftime('%Y-%m-%d')
video.sort_values(['ct','videoname'])[['date','videoname']].drop_duplicates('videoname')
```

```python
video.sort_values(['ct','videoname','viewcnt'],ascending=False).loc[video.ct.dt.strftime('%H:%M')=='21:00'][['videoname','viewcnt','ct']].drop_duplicates('videoname',keep='last').reset_index(drop=True)
```

```python
target=video.sort_values('ct').drop_duplicates('videoname',keep='last')
target['ratio']=target.dislikecnt/target.likecnt
target.sort_values('ratio')[['videoname','ratio']].reset_index(drop=True)
```

```python
target=video.loc[(video.ct>=pd.to_datetime('2021-11-01 00:00:00'))&(video.ct<=pd.to_datetime('2021-11-01 15:00:00'))].reset_index(drop=True)
target[['videoname','viewcnt']].groupby('videoname').max()-target[['videoname','viewcnt']].groupby('videoname').min()
```
